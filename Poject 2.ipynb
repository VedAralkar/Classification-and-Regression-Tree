```python
# Comparing both Deciion Tree Classifier and Random Forest Classifier on scikit-learn's digit dataset

from sklearn.datasets import load_digits
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
# Performing Decision Tree Classifier on scikit-learn’s hand-written digits (each data point is an 8X8 image of a single digit) with tuning the classifier.

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.datasets import load_digits
```

```python
# Load the digits dataset
digits = load_digits()
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2, random_state=42)
```

```python
# Define the parameter grid for hyperparameter tuning
param_grid = {
    'max_depth': [5, 10, 15, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}
```

```python
# Create a decision tree classifier object
dtc = DecisionTreeClassifier()
```

```python
# Create a grid search object
grid_search = GridSearchCV(dtc, param_grid, cv=5)
```

```python
# Fit the grid search object to the training data
grid_search.fit(X_train, y_train)
```

```python
# Get the best hyperparameters
best_params = grid_search.best_params_
```

```python
# Create a new decision tree classifier object with the best hyperparameters
dtc_tuned = DecisionTreeClassifier(**best_params)
```

```python
# Fit the new classifier to the training data
dtc_tuned.fit(X_train, y_train)
```

```python
# Make predictions on the test data
y_pred = dtc_tuned.predict(X_test)
```

```python
# Print the confusion matrix
conf_mat = confusion_matrix(y_test, y_pred)
print(conf_mat)
```

```python
import seaborn as sns

# Plot the confusion matrix heatmap
sns.heatmap(conf_mat, annot=True, cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Comaparing between Decision Tree Regressor and Random Regressor on california Dataset.

from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the California housing dataset
data = fetch_california_housing()
```

```python
# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.2, random_state=42)
```

```python
# Create a Decision Tree Regressor
dt_regressor = DecisionTreeRegressor(random_state=42)
dt_regressor.fit(X_train, y_train)
```

```python
# Create a Random Forest Regressor
rf_regressor = RandomForestRegressor(random_state=42)
rf_regressor.fit(X_train, y_train)
```

```python
# Predict on the test set
y_pred_dt = dt_regressor.predict(X_test)
y_pred_rf = rf_regressor.predict(X_test)
```

```python
# Compute the mean squared error and r2 score for both models
mse_dt = mean_squared_error(y_test, y_pred_dt)
r2_dt = r2_score(y_test, y_pred_dt)
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_rf = r2_score(y_test, y_pred_rf)
```

```python
# Print the mean squared error and r2 score for both models
print("Decision Tree Regressor:")
print("Mean squared error:", mse_dt)
print("Coefficient of determination:", r2_dt)
print("Random Forest Regressor:")
print("Mean squared error:", mse_rf)
print("Coefficient of determination:", r2_rf)
```

```python
import matplotlib.pyplot as plt

# Create a dictionary to store the mean squared error and coefficient of determination values for both models
metrics = {'Decision Tree Regressor': [mse_dt, r2_dt], 'Random Forest Regressor': [mse_rf, r2_rf]}

# Plot the bar graph
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(metrics.keys(), [metric[0] for metric in metrics.values()], yerr=None, align='center', alpha=0.5, ecolor='black', capsize=10, label='Mean Squared Error')
ax.bar(metrics.keys(), [metric[1] for metric in metrics.values()], yerr=None, align='center', alpha=0.5, ecolor='black', capsize=10, label='Coefficient of Determination')
ax.set_ylabel('Value')
ax.set_title('Performance Comparison')
ax.legend()
plt.show()
```

```python
# Printing the feature importance (aka Gini importance) for scikit-learn california-housing dataset for Random Regressor.

from sklearn.datasets import fetch_california_housing
from sklearn.ensemble import RandomForestRegressor
```

```python
housing = fetch_california_housing()
X = housing.data
y = housing.target
```

```python
rf = RandomForestRegressor()
rf.fit(X, y)
```

```python
importances = rf.feature_importances_
for feature, importance in zip(housing.feature_names, importances):
    print(f"{feature}: {importance:.3f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Get feature importance scores
importances = rf.feature_importances_

# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Rearrange feature names so they match the sorted feature importances
names = [housing.feature_names[i] for i in indices]

# Create plot
plt.figure()

# Create plot title
plt.title("Random Forest Regressor Feature Importance")

# Add bars
plt.bar(range(X.shape[1]), importances[indices])

# Add feature names as x-axis labels
plt.xticks(range(X.shape[1]), names, rotation=90)

# Show plot
plt.show()
```
